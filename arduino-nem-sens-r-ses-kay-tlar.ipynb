```python
from gtts import gTTS

# Mesajlar
messages = {
    "dry": "Toprak çok kuru. Lütfen sulama yapınız.",
    "moist": "Toprak nemli. Sulama yapmanıza gerek yok.",
    "wet": "Toprak çok sulu. Daha fazla su eklemeyiniz."
}

# Dilleri belirtiyoruz
language = 'tr'  # Türkçe dil kodu

# Ses dosyalarını oluştur ve kaydet
for key, text in messages.items():
    tts = gTTS(text=text, lang=language)
    filename = f"{key}_soil.mp3"  # Örneğin: dry_soil.mp3
    tts.save(filename)
    print(f"{filename} kaydedildi.")

print("Tüm ses dosyaları başarıyla oluşturuldu.")
```
